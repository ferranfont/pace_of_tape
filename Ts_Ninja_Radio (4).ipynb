```python
import paho.mqtt.client as mqtt
import pandas as pd
import ssl
import json
import os
import datetime
```

```python
# Definir una lista para almacenar los datos
datos = []
df = pd.DataFrame(columns=['Datos'])
headers_written = False
```

```python
# Ruta de los archivos de salida:
path_output = "Ninja_Radio/"
## Nombre del archivo de salida (tanto el csv como el HTML)
```

```python
def on_connect(client, userdata, flags, rc):
    print("Connected with result code "+str(rc))
    client.subscribe("Time-and-sales")

def on_message(client, userdata, msg):
    global df
    datos.append(str(msg.payload))
    
    
    df_new = pd.DataFrame(datos, columns=['Datos'])
    df_new['Datos'] = df_new['Datos'].str.strip("b'")  # Eliminar los caracteres iniciales y finales no deseados
    df_new = pd.json_normalize(df_new['Datos'].apply(json.loads))
    df_new = df_new.iloc[:, 1:]
    df = pd.concat([df, df_new], ignore_index=True)
    datos.clear()
    print(df_new)
    save_data_to_csv()
    
def save_data_to_csv():
    now = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
    nombre_output = "TS_Ninja_Radio_v1"
    path_output = "Output/"
    df.to_csv(f"{path_output}{nombre_output}.csv", sep=";", index=False)
   
    

# Crear el cliente MQTT
client = mqtt.Client("test")
client.on_connect = on_connect
client.on_message = on_message
client.connect("localhost", 7890)
client.loop_forever()

```
